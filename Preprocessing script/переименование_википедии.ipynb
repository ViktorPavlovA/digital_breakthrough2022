```python
subject_name_data = subject_name_data[(subject_name_data['subject_name']!='Севастополь[a]') &
                  (subject_name_data['subject_name']!='Крым[a]') &
                  (subject_name_data['subject_name']!='Архангельская безНАО ')
                                     & (subject_name_data['subject_name']!='Тюменская областьбез ХМАО и ЯНАО')]
subject_name_data['subject_name'] = subject_name_data['subject_name'].str.replace('край', '')
subject_name_data['subject_name'] = subject_name_data['subject_name'].str.replace('область', '')
```

```python
test = pd.read_csv('test.csv', sep=';')
test.head()
```

```python
ones_test = test[test['label']==1].groupby(['hex'])['label'].count()
ones_df_test = pd.DataFrame(ones_test).rename({'label':'ones'}, axis=1).reset_index()

zeros_test = test[test['label']==0].groupby(['hex'])['label'].count()
zeros_df_test = pd.DataFrame(zeros_test).rename({'label':'zeros'}, axis=1).reset_index()

df_temp_test = zeros_df_test.merge(ones_df_test, how='outer', on='hex').fillna(0)
df_temp_test['sum'] = df_temp_test['zeros'] + df_temp_test['ones']

df_temp_test['percent_zeros'] = df_temp_test['zeros'] / df_temp_test['sum']
df_temp_test['percent_ones'] = df_temp_test['ones'] / df_temp_test['sum']

# объединение
merged_test = test.merge(df_temp_test[['hex', 'percent_zeros', 'percent_ones']], how='outer', on='hex')
```
